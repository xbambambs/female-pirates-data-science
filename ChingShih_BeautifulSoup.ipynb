```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
# Load the webpage
r= requests.get("https://en.wikipedia.org/wiki/Zheng_Yi_Sao")

# Convert to a beautiful soup object
soup = bs(r.content)

# Print HTML
contents = soup.prettify()
print(contents)
```

```python
info_box = soup.find(class_="infobox biography vcard")
info_rows = info_box.find_all("tr")
for rows in info_rows:
    print(row.prettify())
```

```python
pirate_info = {}

for index, row in enumerate(info_rows):
    if index ==0:
        pirate_info["Name"] = row.find("th").get_text()
        
print(pirate_info)
```

```python
pirate_info = {}

for index, row in enumerate(info_rows):
    if index ==0:
        pirate_info["Name"] = row.find("th").get_text()
    elif index == 1:
        continue
    else:
        content_key = row.find("td").get_text()
        content_value = row.find("row")
        pirate_info[content_key] = content_value
        
        
print(pirate_info)
```
